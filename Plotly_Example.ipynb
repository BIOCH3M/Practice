```python
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly.graph_objects as go
np.random.seed(42)
```

```python
rx=np.random.randint(1,101,100)
ry=np.random.randint(1,101,100)
```

### Scatter

```python
data=[go.Scatter(x=rx, y=ry, mode='markers', marker={'size':12,
                                                    'color':'rgb(51,205,153)',
                                                    'symbol':'pentagon',
                                                    'line': {'width':2}})]
layout=go.Layout(title='Hello Plot', xaxis={'title':'MY X'},yaxis={'title':'My Y'}, hovermode='closest')

fig=go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

### Linechart

```python
rx=np.linspace(0,1,100)
ry=np.random.randn(100)
trace0=go.Scatter(x=rx, y=ry+5, mode='markers', name='markers' )
trace1=go.Scatter(x=rx, y=ry, mode='lines', name='mylines' )
trace2=go.Scatter(x=rx, y=ry-2, mode='lines+markers', name='mylines+markers' )
data=[trace0, trace1, trace2]
layout=go.Layout(title='Line Chart')
fig=go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

### Bar Chart

```python
url='https://raw.githubusercontent.com/Pierian-Data/Plotly-Dashboards-with-Dash/master/Data/2018WinterOlympics.csv'
df = pd.read_csv(url, error_bad_lines=False)

df.head()

```

```python
data=[go.Bar(x=df['NOC'], y=df['Total'])]
layout=go.Layout(title="olympics")
fig=go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

```python
trace1=go.Bar(x=df['NOC'], y=df['Gold'], name='Gold', marker={'color':'#FFD700'})
trace2=go.Bar(x=df['NOC'], y=df['Silver'], name='Silver', marker={'color':'#9EA0A1'})
trace3=go.Bar(x=df['NOC'], y=df['Bronze'], name='Bronze', marker={'color':'#CD7F32'})

data=[trace1, trace2, trace3]
layout=go.Layout(title="olympics", barmode='stack')
fig=go.Figure(data=data, layout=layout)
pyo.plot(fig)
```
