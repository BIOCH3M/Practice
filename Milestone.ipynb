```python
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly.graph_objects as go
import dash
import dash_core_components as dcc
import dash_html_components as html
import dash_auth
from dash.dependencies import Input, Output, State
import pandas_datareader as pdr
from datetime import datetime
```

```python
USERNAME_PASSWORD_PAIRS=[['username','password'],['jamesbond','007']]



app=dash.Dash()
auth=dash_auth.BasicAuth(app,USERNAME_PASSWORD_PAIRS)
nsdq=pd.read_csv('https://raw.githubusercontent.com/Pierian-Data/Plotly-Dashboards-with-Dash/master/Data/NASDAQcompanylist.csv')
nsdq.set_index('Symbol', inplace=True)
options=[]

for tic in nsdq.index:
    mydict={}
    mydict['label']=nsdq.loc[tic]['Name']
    mydict['value']=tic
    options.append(mydict)



app.layout=html.Div([
    html.Div([html.H1('Stock Ticker Dashboard'),
              html.H3('Enter Stock Symbol', style={'paddingRight':'30px'}),  
              dcc.Dropdown(id='my_stock_picker', options=options,value=['TSLA'], multi=True)],
             style={'display':'inline-block', 'verticalAlign':'top', 'width':'30%'}),
    html.Div([html.H3('Select Period'),
              #1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
              dcc.Dropdown(id='my_date_picker',options=[{'label':'1 Day','value':'1d'},
                                                        {'label':'5 Day','value':'5d'},
                                                        {'label':'1 Month','value':'1m'},
                                                        {'label':'3 Month','value':'3m'},
                                                        {'label':'6 Month','value':'6m'},
                                                        {'label':'1 Year','value':'1y'},
                                                        {'label':'5 Year','value':'5y'},
                                                        {'label':'10 Year','value':'10y'},
                                                        {'label':'Year TO Day','value':'ytd'},
                                                        {'label':'MAX','value':'max'}],
                value='max')
             ], style={'display':'inline-block'}),
    html.Div([
        html.Button(id='submit_button',
                   n_clicks=0,
                   children='Submit',
                   style={'fontSize':24, 'marginLeft':'30px'})
    ], style={'display':'inline-block'}),
    dcc.Graph(id='my_graph',
              figure={'data':[{'x':[1,2], 'y':[3,1]}],
                      'layout':{'title':'Default Title'}})
], style={'verticalAlign':'top'})




@app.callback(Output('my_graph', 'figure'),
              [Input('submit_button', 'n_clicks')],
             [State('my_stock_picker','value'),
              State('my_date_picker','value')])

def update_graph(n_clicks,stock_ticker,my_date_picker):
    
    import yfinance as yf 
    
    traces=[]
    for tic in stock_ticker:
        df=yf.Ticker(tic).history(period=my_date_picker)
        traces.append({'x':df.index,'y':df['Close'], 'name':tic})
        fig={'data':traces,
            'layout':{'title':stock_ticker}


            }
    return fig




if __name__=='__main__':
    app.run_server()
    
    
    
    
```
